```python
import inspect

def inspectImage(image):
    print("The type of the image is " + str(type(image)))
    inspect.getmro(type(image))
    print('{}px x {}px'.format(image.width, image.height))

inspectImage(image)
```

```python
from PIL import ImageFilter

def generateFilteredImage(image, filter):
    return image.filter(filter)

image = image.convert('RGB')
generateFilteredImage(image, PIL.ImageFilter.BLUR)
```

```python
generateFilteredImage(image, PIL.ImageFilter.EMBOSS)
```

```python
generateFilteredImage(image, PIL.ImageFilter.CONTOUR)
```
